```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
import numpy as np
from numpy import linalg
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tokenize import PunktSentenceTokenizer
from nltk.corpus import webtext
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('vader_lexicon')
```

```python
with open('kindle.txt', encoding ='ISO-8859-2') as f:
    text = f.read()

text = text.lower()
```

```python
sent_tokenizer = PunktSentenceTokenizer(text)
sents = sent_tokenizer.tokenize(text)

print(word_tokenize(text))
print(sent_tokenize(text))
```

```python
#Stem
porter_stemmer = PorterStemmer()

nltk_tokens = word_tokenize(text)

for w in nltk_tokens:
    print(f"Actual: {w} Stem: {porter_stemmer.stem(w)}")
```

```python
#Lemma
word_lemma = WordNetLemmatizer()

for w in nltk_tokens:
    print(f"Actual: {w} Lemma: {word_lemma.lemmatize(w)}")
```

```python
tagged_tokens = nltk.pos_tag(nltk_tokens)
print(tagged_tokens)
```

```python
#Mejor version de lemma
from nltk.corpus import wordnet

for word, tag in tagged_tokens:
    first_letter = tag[0].lower()

    if first_letter == 'n':
        pos_wordnet = wordnet.NOUN
    elif first_letter == 'v':
        pos_wordnet = wordnet.VERB
    elif first_letter == 'j':
        pos_wordnet = wordnet.ADJ
    elif first_letter == 'r':
        pos_wordnet = wordnet.ADV
    else:
        pos_wordnet = wordnet.NOUN

    print(f"Actual: {word} Lemma: {word_lemma.lemmatize(word, pos=pos_wordnet)}")
```

```python
sentiment_model = SentimentIntensityAnalyzer()

with open('kindle.txt') as f:
    for text in f.read().split("\n"):
        print(text)
        scores = sentiment_model.polarity_scores(text)
        for key in sorted(scores):
            print(f"{key}: {scores[key]}, ")
```
