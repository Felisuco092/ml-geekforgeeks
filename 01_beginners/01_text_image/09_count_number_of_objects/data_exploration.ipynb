Primero cargamos la librerias y la imagen

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('bitss.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

Quitamos el ruido con un suavizado gaussiano

```python
blur = cv2.GaussianBlur(gray_image, (11, 11), 5)
plt.imshow(blur, cmap='gray')
```

Utilizando la tecnica de Canny para detectar contornos con threshold maximos y minimos, buscamos los edges

```python
canny = cv2.Canny(blur, 70, 110)
plt.imshow(canny, cmap='gray')
```

Los edges no estan conectados, vamos a conectarlos con dilate

```python
dilated = cv2.dilate(canny, (5,5), iterations=4)
#Lo siguiente puedes comentarlo es para ver como seria si no hacemos dilate
#dilated = canny
plt.imshow(dilated, cmap='gray')
```

Buscamos contornos y los dibujamos

```python
(cnt, hierarchy) = cv2.findContours(dilated.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb,cnt,-1,(0,255,0),2)

plt.imshow(rgb)
```
