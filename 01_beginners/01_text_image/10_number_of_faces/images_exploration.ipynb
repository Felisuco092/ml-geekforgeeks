## En esta libreta usaremos opencv para manejar las imagenes, dlib para detectar caras y numpy para operaciones con arrays.

```python
# Import libraries
import cv2
import numpy as np
import dlib
```

```python
#Cogemos la camara del pc
cap = cv2.VideoCapture(0)
```

```python
# Cargamos el modelo de detección de rostros de dlib
detector = dlib.get_frontal_face_detector()
```

```python
#Aqui capturamos los frames de la camara y los mostramos en una ventana la cara
while True:
    # Capturamos un frame de la cámara
    ret, frame = cap.read()

    # Convertimos el frame a escala de grises
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Detectamos los rostros en el frame
    faces = detector(gray)

    # Iterador de caras
    i = 0
    for face in faces:
         # Coordenadas de cara esquina superior izquierda y esquina inferior derecha
        x1, y1 = face.left(), face.top()
        x2, y2 = face.right(), face.bottom()

        # Dibujamos un rectángulo alrededor de la cara
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

        # Mostramos el número de caras detectadas en la esquina superior izquierda
        i += 1
        cv2.putText(frame, f'face num {i}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        print(f'face num {i} detected at coordinates: ({x1}, {y1}), ({x2}, {y2})')

    # Mostramos el frame con las caras detectadas
    cv2.imshow('Face Detection', frame)
    
    # Para salir dandole a la 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberamos la cámara y cerramos las ventanas
cap.release()
cv2.destroyAllWindows()


```
