Primero cargamos nuestra imagen y la pasamos a escala de grises

```python
import cv2

img = cv2.imread('input_image.png')

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img = cv2.medianBlur(gray_img, 5)
```

Detectamos Edges atraves de un adaptativo threshold

```python
import matplotlib.pyplot as plt
edges = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
edges_2 = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 5) #Este esta para hacer pruebas, no es el que se usa en el proyecto final
plt.subplot(1, 2, 1)
plt.imshow(edges, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(edges_2, cmap='gray')
```

Hacemos que el color sea mas suave

```python
filter = cv2.bilateralFilter(img, 9, 250, 250)
plt.imshow(filter)
```

Y los combinamos

```python
cartoon = cv2.bitwise_and(filter, filter, mask=edges)
plt.imshow(cartoon)
```
